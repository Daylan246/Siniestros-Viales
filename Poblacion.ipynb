```python
# Eliminamos la primer fila que contiene el tÃ­tulo de las columnas
df_poblacion = df_poblacion.drop(df_poblacion.index[0])

# Limpieza de la columna 'Poblacion 2022' antes de convertir a int
df_poblacion['Poblacion 2022'] = df_poblacion['Poblacion 2022'].replace('\xa0', '', regex=True)

# Convertimos la columna 'Poblacion 2022' a int despuÃ©s de la limpieza
df_poblacion['Poblacion 2022'] = df_poblacion['Poblacion 2022'].astype(int)

# PoblaciÃ³n total de 2022
censo_2022 = 3120612

# Tasa de crecimiento total entre 2010 y 2022
tasa_crecimiento_total = 7.97 / 100

# Calcular la tasa de crecimiento anual
tasa_crecimiento_anual = (1 + tasa_crecimiento_total) ** (1 / 12) - 1

# Crear una nueva columna con los valores estimados para 2021
df_poblacion['Poblacion 2021'] = df_poblacion['Poblacion 2022'] / (1 + tasa_crecimiento_anual)
```

```python

# Datos de poblaciÃ³n de 2010 y tasas de crecimiento para nuestro calculo 
data = {
    'Comuna': ['Comuna 13', 'Comuna 14', 'Comuna 12', 'Comuna 4', 'Comuna 1', 'Comuna 7', 'Comuna 8', 'Comuna 6', 'Comuna 11', 'Comuna 15', 'Comuna 3', 'Comuna 5', 'Comuna 10', 'Comuna 9', 'Comuna 2'],
    'Poblacion 2010': [231331, 225970, 200116, 218245, 205886, 220591, 187237, 176076, 189832, 182574, 187537, 179005, 166022, 161797, 157932],
    'Crecimiento 2010-2022': [14.64, 10.19, 18.07, 5.81, 8.44, -1.70, 9.40, 15.73, 7.19, 8.39, 4.22, 8.29, 3.47, 4.78, 0.27]
}

df_poblacion = pd.DataFrame(data)

# Convertir tasas de crecimiento en factores de crecimiento
df_poblacion['Factor de crecimiento'] = 1 + df_poblacion['Crecimiento 2010-2022'] / 100

# Calcular poblaciÃ³n para cada aÃ±o
for year in range(2016, 2022):
    n = year - 2010
    df_poblacion[f'Poblacion {year}'] = df_poblacion['Poblacion 2010'] * df_poblacion['Factor de crecimiento']**(n/12)
```
